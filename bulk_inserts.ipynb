```python
from config import DATABASE_URI

import psycopg2 as pg
from psycopg2.extras import execute_values
import pandas as pd
import numpy as np
```

```python
# Tickers table insert
tickers = pd.read_csv("tickers.csv")
tickers = tickers.dropna(subset=["permaticker", "ticker"])
tickers = tickers.drop_duplicates(["permaticker", "ticker"])
tickers["isdelisted"] = tickers["isdelisted"].replace({"N": "false", "Y": "true"})
tickers = tickers.replace({np.nan: None})
tickers = tickers.drop(columns=["table"])

sql = f"INSERT INTO tickers ({', '.join(tickers.columns)}) VALUES %s"

with pg.connect(DATABASE_URI) as conn:
    with conn.cursor() as cur:
        execute_values(cur, sql, list(tickers.itertuples(index=False, name=None)))
```

```python
# Equity prices table insert
conn = pg.connect(DATABASE_URI)
with conn:
    with conn.cursor() as cur:
        cur.execute("SELECT permaticker, ticker FROM tickers")
        results = cur.fetchall()
    # Get all tickers we have defined in the tickers table
    valid_tickers = pd.DataFrame(results, columns=["permaticker", "ticker"])
    
with pd.read_csv("sep.csv", chunksize=100000) as reader:
    for chunk in reader:
        chunk = chunk.merge(valid_tickers, on="ticker", how="left")
        chunk["frequency"] = "DAILY"
        chunk = chunk.replace({np.nan: None})
        chunk = chunk.dropna(subset=["ticker"])

        sql = f"INSERT INTO prices ({', '.join(chunk.columns)}) VALUES %s ON CONFLICT (ticker, date, frequency) DO UPDATE SET ({', '.join(chunk.columns)}) = ({', '.join('EXCLUDED.' + chunk.columns)})"

        with conn:
            with conn.cursor() as cur:
                execute_values(cur, sql, list(chunk.itertuples(index=False, name=None)))

        print("Chunk Complete")

conn.close()
```

```python
# Fundamentals table insert
conn = pg.connect(DATABASE_URI)
with conn:
    with conn.cursor() as cur:
        cur.execute("SELECT permaticker, ticker FROM tickers")
        results = cur.fetchall()
    # Get all tickers we have defined in the tickers table
    valid_tickers = pd.DataFrame(results, columns=["permaticker", "ticker"])
    
with pd.read_csv("sf1.csv", chunksize=100000) as reader:
    for chunk in reader:
        # Select only rows with valid tickers, this avoids foreign key constraint violations
        chunk = chunk.merge(valid_tickers, on="ticker", how="left")
        chunk = chunk.replace({np.nan: None})
        chunk = chunk.dropna(subset=["ticker"])

        sql = f"INSERT INTO fundamentals ({', '.join(chunk.columns)}) VALUES %s ON CONFLICT (ticker, dimension, datekey, reportperiod) DO UPDATE SET ({', '.join(chunk.columns)}) = ({', '.join('EXCLUDED.' + chunk.columns)})"

        with conn:
            with conn.cursor() as cur:
                execute_values(cur, sql, list(chunk.itertuples(index=False, name=None)))

        print("Chunk Complete")

conn.close()
```
